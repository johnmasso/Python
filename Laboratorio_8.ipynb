```python

#EJERCICIOS PARTE 1
#EJERCICIO 1
# Importar bibliotecas necesarias
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Crear el Diccionario con los datos proporcionados
data = {
    'Producto': ['Producto A', 'Producto B', 'Producto C', 'Producto D', 'Producto E', 'Producto F', 'Producto G', 'Producto H'],
    'Bogotá': [150, 100, 90, 120, 160, 130, 110, 95],
    'Cali': [200, 180, 150, 170, 210, 175, 140, 160],
    'Bucaramanga': [250, 220, 300, 280, 230, 240, 210, 200]
}

# Convertir el diccionario en un DataFrame
df = pd.DataFrame(data)

# Configuración del estilo de Seaborn
sns.set(style="whitegrid", palette="pastel")

# Crear la figura y el gráfico de barras
plt.figure(figsize=(10, 6))
df_melted = df.melt(id_vars='Producto', var_name='Ciudad', value_name='Ventas')

# Graficar los datos
sns.barplot(data=df_melted, x='Producto', y='Ventas', hue='Ciudad')
plt.title('Ventas de Productos en Diferentes Ciudades', fontsize=16)
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.legend(title='Ciudad')
plt.xticks(rotation=45)

# Mostrar la gráfica
plt.show()
```

```python
#EJERCICIO 2
# Importar bibliotecas necesarias
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Crear el Diccionario con los datos proporcionados
data = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto'],
    'Papa (Cali)': [100, 150, 200, 250, 300, 400, 300, 150],
    'Cebolla (Bogotá)': [90, 120, 160, 200, 240, 100, 200, 150],
    'Tomate (Cartagena)': [80, 110, 150, 190, 230, 330, 350, 240]
}

# Convertir el diccionario en un DataFrame
df = pd.DataFrame(data)

# Configuración del estilo de Seaborn
sns.set(style="whitegrid", palette="muted")

# Crear la figura y los gráficos de línea
plt.figure(figsize=(12, 6))

# Graficar cada producto en función del mes
sns.lineplot(data=df, x='Mes', y='Papa (Cali)', marker='o', label='Papa (Cali)', color='green')
sns.lineplot(data=df, x='Mes', y='Cebolla (Bogotá)', marker='o', label='Cebolla (Bogotá)', color='orange')
sns.lineplot(data=df, x='Mes', y='Tomate (Cartagena)', marker='o', label='Tomate (Cartagena)', color='red')

# Personalizar el gráfico
plt.title('Evolución de las Ventas de Productos en Diferentes Ciudades', fontsize=16)
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.legend(title='Producto (Ciudad)')
plt.xticks(rotation=45)

# Mostrar la gráfica
plt.show()
```

```python
#EJERCICIOS PARTE 2
#EJERCICIO 1
# Importar bibliotecas necesarias
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Crear el diccionario con los datos proporcionados
data = {
    'Puntuacion': [50, 60, 65, 70, 75, 78, 80, 82, 85, 87, 90, 92, 94, 96, 98, 99, 100, 102, 103, 105],
    'Nivel_de_Motivacion': [
        'Baja', 'Media', 'Media', 'Alta', 'Alta', 'Media', 'Media', 'Alta',
        'Alta', 'Baja', 'Alta', 'Alta', 'Media', 'Alta', 'Alta', 'Media',
        'Alta', 'Media', 'Baja', 'Alta'
    ]
}

# Convertir el diccionario en un DataFrame
df = pd.DataFrame(data)

# Configurar el estilo de Seaborn
sns.set(style="whitegrid")

# Crear un gráfico de histogramas separados con transparencia
plt.figure(figsize=(12, 6))
sns.histplot(
    data=df,
    x='Puntuacion',
    hue='Nivel_de_Motivacion',
    multiple='layer',    # Muestra los histogramas superpuestos sin apilar
    palette='bright',   # Paleta de colores para mayor atractivo
    edgecolor='black',   # Color del borde de las barras
    binwidth=5,          # Ancho de cada bin
    alpha=0.5            # Transparencia para evitar ocultar datos superpuestos
)

# Personalizar el gráfico
plt.title('Distribución de Puntuaciones en el Examen según el Nivel de Motivación', fontsize=16)
plt.xlabel('Puntuación')
plt.ylabel('Frecuencia')

# Mostrar el gráfico
plt.show()

```

```python
# segunda opción del ejercicio 1 con gráficos KDE
# Importar bibliotecas necesarias
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Crear el diccionario con los datos proporcionados
data = {
    'Puntuacion': [50, 60, 65, 70, 75, 78, 80, 82, 85, 87, 90, 92, 94, 96, 98, 99, 100, 102, 103, 105],
    'Nivel_de_Motivacion': [
        'Baja', 'Media', 'Media', 'Alta', 'Alta', 'Media', 'Media', 'Alta',
        'Alta', 'Baja', 'Alta', 'Alta', 'Media', 'Alta', 'Alta', 'Media',
        'Alta', 'Media', 'Baja', 'Alta'
    ]
}

# Convertir el diccionario en un DataFrame
df = pd.DataFrame(data)

# Configurar el estilo de Seaborn
sns.set(style="whitegrid")

# Crear un gráfico de densidad
plt.figure(figsize=(12, 6))
sns.kdeplot(
    data=df,
    x='Puntuacion',
    hue='Nivel_de_Motivacion',
    palette='pastel',   # Paleta de colores
    fill=True,           # Rellena el área bajo las curvas
    alpha=0.5            # Transparencia
)

# Personalizar el gráfico
plt.title('Distribución de Puntuaciones en el Examen según el Nivel de Motivación', fontsize=16)
plt.xlabel('Puntuación')
plt.ylabel('Densidad')

# Mostrar el gráfico
plt.show()
```

```python

# EJERCICIO 2
# Importar bibliotecas necesarias
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Crear el diccionario con los datos proporcionados
data = {
    'Mes': [
        'Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio',
        'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'
    ],
    'SUV': [120, 130, 135, 140, 150, 160, 170, 180, 175, 185, 190, 200],
    'Sedan': [100, 90, 95, 110, 120, 130, 140, 145, 150, 160, 170, 180],
    'Hatchback': [80, 85, 90, 95, 100, 105, 110, 115, 120, 125, 130, 135]
}

# Convertir el diccionario en un DataFrame
df = pd.DataFrame(data)

# Configurar el estilo de Seaborn
sns.set(style="whitegrid")

# Crear el gráfico de líneas
plt.figure(figsize=(12, 6))
plt.plot(df['Mes'], df['SUV'], marker='o', color='b', label='SUV', linewidth=2)
plt.plot(df['Mes'], df['Sedan'], marker='o', color='g', label='Sedan', linewidth=2)
plt.plot(df['Mes'], df['Hatchback'], marker='o', color='r', label='Hatchback', linewidth=2)

# Personalizar el gráfico
plt.title('Ventas Mensuales de Vehículos por Tipo en el Año', fontsize=16)
plt.xlabel('Mes')
plt.ylabel('Número de Ventas')
plt.xticks(rotation=45)  # Rotar los nombres de los meses para mejor lectura
plt.legend(title='Tipo de Vehículo')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)  # Agregar una cuadrícula ligera

# Mostrar el gráfico
plt.show()
```
